```python
import matplotlib.pyplot as plt
%matplotlib inline

fname = "C:\\Personal\\Junior Year\\SES 330\\Lab More Diodes\\Ripple.txt"

X, Y = [], []
for line in open(fname, 'r'):
  values = [float(s) for s in line.split()]
  X.append(values[0])
  Y.append(values[1])  
  


plt.plot(X, Y, 'b-') 
plt.ylabel ('Voltage [Volts]')
plt.xlabel ('Time [ns]')
plt.axhline(y = 0, color = 'black', linestyle = '--')
plt.title('Time vs. Voltage forFull Wave\nBridge Rectifier with Capacitor')

plt.show()
```
