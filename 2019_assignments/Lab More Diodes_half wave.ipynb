```python
import matplotlib.pyplot as plt
%matplotlib inline

fname = "C:\\Personal\\Junior Year\\SES 330\\Lab More Diodes\\Half_Wave.txt"

X, Y, Z = [], [], []
for line in open(fname, 'r'):
  values = [float(s) for s in line.split()]
  X.append(values[0])
  Y.append(values[1])
  Z.append(values[2])  
  


plt.plot(X, Y, label='Input') 
plt.plot(X, Z, label='Output')
plt.ylabel ('Voltage [Volts]')
plt.xlabel ('Time [ns]')
plt.axhline(y = 0, color = 'black', linestyle = '--')
plt.title('Time vs. Voltage for\nHalf Wave Bridge Rectifier')
plt.legend(loc=9, bbox_to_anchor=(1.15, 0.7))

plt.show()
plt.show()
```
