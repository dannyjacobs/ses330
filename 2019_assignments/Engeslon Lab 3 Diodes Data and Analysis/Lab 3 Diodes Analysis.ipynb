```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t1, inputamp, t2, outputamp = np.loadtxt('./Half Wave Rectifier.txt', unpack = True)
```

```python
t = t1 - t1[0]
```

```python
plt.plot(t,inputamp, 'g-')
plt.plot(t,outputamp, '-' )
plt.xlabel('Time (Sec)')
plt.ylabel('Voltage (V)')
plt.title('Half Wave Rectifier')
```
