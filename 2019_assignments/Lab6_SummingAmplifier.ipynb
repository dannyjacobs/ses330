```python
import matplotlib.pyplot as plt
import numpy as np
# if using a jupyter notebook
%matplotlib inline 

x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)
z = (np.sin(x)*-0.9)+2
w = (np.sin(x))+2


plt.plot(x,y,'y-', label='Input')
plt.plot(x,z,'b-', label='Output with op-amp')
plt.plot(x,w,'r-', label='Output without op-amp')



plt.xlabel('Time [ns]')   # string must be enclosed with quotes '  '
plt.ylabel('Voltage [V]')
plt.title('Summing Amplifier\nVoltages vs.time')
plt.legend(loc=9, bbox_to_anchor=(1.3, 0.7))

plt.show()
```
