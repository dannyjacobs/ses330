<h3>1.2 Displaying Data</h3>

<h4>Data from bode analyzer is collected and stored in CSV file. Code below is inputted in order to recall the data collected and plot the visual peak of the resonant frequency.</h4>

```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt("bodeplot.csv", dtype='float', skiprows=3)
```

```python
plt.figure(3)
plt.plot(data[:,0],data[:,1])
plt.title("Frequency vs Gain")
plt.xlabel("Frequency[Hz]")
plt.ylabel("Gain")
```
