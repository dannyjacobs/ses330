```python
import matplotlib.pyplot as plt

fname = "C:\\Program Files\\Git\\ses330\\SES 330 Tran Engelson LPF.txt"

X, Y = [], []
for line in open(fname, 'r'):
  values = [float(s) for s in line.split()]
  X.append(values[0])
  Y.append(values[1])

plt.plot(X, Y)
plt.ylabel ('Gain [dB]')
plt.xlabel ('Frequency [Hz]')
plt.title('Frequency vs. Gain for\nLow Pass Filter')
plt.show()
```
