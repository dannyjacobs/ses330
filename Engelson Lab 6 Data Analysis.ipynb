```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-5,6,0.01)
```

```python
inpt1 = 1.066 * np.sin(2*np.pi*x)
otpt1 = 1.105 * np.sin(2*np.pi*x)+0.02
```

```python
plt.plot(x,inpt1,color='green')
plt.plot(x,otpt1)
plt.xlabel('Time (Sec)')
plt.ylabel('Voltage (V)')
plt.title('Follower Circuit')
plt.axis([-1,1,-1.25,1.25])
plt.axvline(x=0, color = 'red')
plt.axhline(y=0, color = 'red')
```

```python
inpt2 = 1.107 * np.sin(2*np.pi*x)
otpt2 = 10.940 * np.sin(2*np.pi*x)+0.02
```

```python
plt.plot(x,inpt2,color='green')
plt.plot(x,otpt2)
plt.xlabel('Time (Sec)')
plt.ylabel('Voltage (V)')
plt.title('Non-Inverting Circuit')
plt.axis([-1,1,-12,12])
plt.axvline(x=0, color = 'red')
plt.axhline(y=0, color = 'red')
```

```python
inpt3 = 1.021 * np.sin(2*np.pi*x)
otpt3 = 9.413 * -np.sin(2*np.pi*x)+0.02
```

```python
plt.plot(x,inpt3,color='green')
plt.plot(x,otpt3)
plt.xlabel('Time (Sec)')
plt.ylabel('Voltage (V)')
plt.title('Inverting Circuit')
plt.axis([-1,1,-12,12])
plt.axvline(x=0, color = 'red')
plt.axhline(y=0, color = 'red')
```

```python
inpt4 = 1.066 * np.sin(2*np.pi*x)
otpt4 = 0.935 * np.sin(2*np.pi*x)+1.02
```

```python
plt.plot(x,inpt4,color='green')
plt.plot(x,otpt4)
plt.xlabel('Time (Sec)')
plt.ylabel('Voltage (V)')
plt.title('Offset Component Only')
plt.axis([-1,1,-5,5])
plt.axvline(x=0, color = 'red')
plt.axhline(y=0, color = 'red')
```

```python
inpt5 = 1.066 * np.sin(2*np.pi*x)
otpt5 = 0.935 * -np.sin(2*np.pi*x)+0.02
```

```python
plt.plot(x,inpt5,color='green')
plt.plot(x,otpt5)
plt.xlabel('Time (Sec)')
plt.ylabel('Voltage (V)')
plt.title('Inverter Component Only')
plt.axis([-1,1,-5,5])
plt.axvline(x=0, color = 'red')
plt.axhline(y=0, color = 'red')
```
