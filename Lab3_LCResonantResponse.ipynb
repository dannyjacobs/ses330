```python
import matplotlib.pyplot as plt
%matplotlib inline

fname = "C:\\Personal\\Junior Year\\SES 330\\Lab 3\\Lab3_LC_SinusoidResponse.txt"

X, Y = [], []
for line in open(fname, 'r'):
  values = [float(s) for s in line.split()]
  X.append(values[0])
  Y.append(values[1])
  


plt.plot(X, Y)
plt.ylabel ('Phase Difference [degrees]')
plt.xlabel ('Frequency [Hz]')
plt.axhline(y = 0, color = 'black', linestyle = '--')
plt.axvline(x = 15915, color = 'green', linestyle = '--')
plt.axvline(x = 9999, color = 'black', linestyle = '--')
plt.title('Frequency vs. Phase Difference for\na LC Resonant Circuit')
plt.show()
```
