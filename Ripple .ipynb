```python
import matplotlib.pyplot as plt
import numpy as np

fname = "C:\\Users\\katko\\Documents\\GitHub\\ses330\\Lab Data\\Part2Lab3-DCRipple.txt"

dT = []
y = []
z = []
y,z = np.loadtxt(fname, dtype=None, skiprows = 5, usecols=(2,5),unpack=True)
dT = np.genfromtxt(fname, dtype=None, skip_header = 5, encoding=None, usecols=(1), unpack=True)

fig,ax = plt.subplots()
ax.plot(dT,y)
ax.set(xlabel='Time(s)',ylabel='Voltage(V)',title='Time vs Voltage')
ax.grid()
plt.show()

fig,ax = plt.subplots()
ax.plot(dT,z)
ax.set(xlabel='Time(s)',ylabel='Voltage(V)',title='Time vs Voltage')
ax.grid()
plt.show()
```
