```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f, g, p = np.loadtxt('./Engelson Tran RLC Band Pass 100k.txt', skiprows = 3, unpack = True)
```

```python
plt.plot(f,g, '.-')
plt.xlabel('Freq (Hz)')
plt.ylabel('Gain (dB)')
plt.title('Band Pass Filter 100k')
```

```python
plt.plot(f,g, '.-')
plt.xlabel('Freq (Hz)')
plt.ylabel('Gain (dB)')
plt.title('Band Pass Filter 100k')
plt.axis([16000,17200,-20,-12])
```

```python
three_dB = max(g) - 3
```

```python
plt.plot(f,g, '.-')
plt.xlabel('Freq (Hz)')
plt.ylabel('Gain (dB)')
plt.title('Band Pass Filter 100k 3dB Point')
plt.axhline(y = three_dB, color = 'red', linestyle = '--')
plt.axis([16000,17200,-20,-12])
plt.axvline(x = 16304, color = 'green', linestyle = '--')
plt.axvline(x = 17020, color = 'green', linestyle = '--')
```
