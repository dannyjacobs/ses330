# Import numpy and matplotlib

```python
import numpy as np
import matplotlib.pyplot as pl
```

# Case 1 Input 100mV 100Hz, Collector 4V, VCC 2V

## Import data set

```python
D = np.genfromtxt('lab5.csv', comments='#', skip_header=28, names=True, delimiter = ',')
```

```python
D.dtype
```

```python
pl.figure()
pl.plot(D['Time_s'], D['Channel_1_V'], label='Input')
pl.plot(D['Time_s'], D['Channel_2_V'], label='Output')
pl.xlabel('time (s)')
pl.ylabel('voltage (v)')
pl.legend()
pl.title ('Input 100mV 100Hz, Collector 4V, VCC 2V')
```

```python
fig, ax1 = pl.subplots()

color = 'red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('voltage (v)', color=color)
ax1.plot(D['Time_s'], D['Channel_1_V'], label='Input', color=color)
ax1.tick_params(axis='y', labelcolor=color)
pl.legend(title='legend', bbox_to_anchor=(1.3,1), loc='upper left')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'blue'
ax2.set_ylabel('voltage (v)', color=color)  # we already handled the x-label with ax1
ax2.plot(D['Time_s'], D['Channel_2_V'], label='Output', color=color)
ax2.tick_params(axis='y', labelcolor=color)
pl.legend(title='legend', bbox_to_anchor=(1.3,1), loc='lower left')

pl.title ('Input 100mV 100Hz, Collector 4V, VCC 2V')

pl.show()
```
