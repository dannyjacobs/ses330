This code sets up python with numpy and the plotting program. It then assigns the CSV files to variables.

```python
import numpy as np
import matplotlib.pyplot as pl
A = np.genfromtxt('6 l 2 open loop.csv', comments='#', skip_header=23, names=True, delimiter=',')
B = np.genfromtxt('6 l 3 1 voltage divider.csv', comments='#', skip_header=24, names=True, delimiter=',')
C = np.genfromtxt('6 l 3 1 voltage divider 3db.csv', comments='#', skip_header=25, names=True, delimiter=',')
D = np.genfromtxt('6 l 3 1 voltage divider 3db math.csv', comments='#', skip_header=24, names=True, delimiter=',')
E=np.genfromtxt('6 l 3 closed loop follower.csv', comments='#', skip_header=24, names=True, delimiter=',')
F=np.genfromtxt('6 l 3 2 1k before feedback.csv', comments='#', skip_header=25, names=True, delimiter=',')
G=np.genfromtxt('6 l 3 2 1k after feedback.csv', comments='#', skip_header=25, names=True, delimiter=',')
H=np.genfromtxt('6 l 4 10k and 1k amp.csv', comments='#', skip_header=24, names=True, delimiter=',')
I=np.genfromtxt('6 l 4 10k and 1k amp network.csv', comments='#',skip_header=25,  names=True, delimiter=',')
J=np.genfromtxt('6 l 5 10k and 1k amp inverse.csv', comments='#', skip_header=24, names=True, delimiter=',')
K=np.genfromtxt('6 l 5 10k and 1k amp inverse network.csv', comments='#', skip_header=34, names=True, delimiter=',')
```

This code plots the variables, and sets up the plot with lables

```python
A.dtype
A['Time_s'], A['Channel_1_V'], A['Channel_2_V']
pl.figure()
pl.plot(A['Time_s'], A['Channel_1_V'], label='Input')
pl.plot(A['Time_s'], A['Channel_2_V'], label='Output')
pl.xlabel('Time (S)')
pl.ylabel('Voltage (V)')
pl.legend()
pl.title('Open Loop Test Circuit')

B.dtype
B['Time_s'], B['Channel_1_V'], B['Channel_2_V']
pl.figure()
pl.plot(B['Time_s'], B['Channel_1_V'], label='Input')
pl.plot(B['Time_s'], B['Channel_2_V'], label='Output')
pl.xlabel('Time (s)')
pl.ylabel('Voltage (V)')
pl.legend()
pl.title('Voltage Divider Op-Amp')

C.dtype
C['Frequency_Hz'], C['Channel_1_Magnitude_dB'], C['Channel_2_Magnitude_dB'],
pl.figure()
pl.plot(C['Frequency_Hz'], C['Channel_1_Magnitude_dB'], label='Input')
pl.plot(C['Frequency_Hz'], C['Channel_2_Magnitude_dB'], label='Output')
pl.xlabel('Frequency (Hz)')
pl.ylabel('Magnitude (dB)')
pl.legend()
pl.title('Voltage Divider 3dB')

D.dtype
D['Time_s'], D['Channel_1_V'], D['Channel_2_V'], D['Math_1_V']
pl.figure()
pl.plot(D['Time_s'], D['Channel_1_V'], label='Input')
pl.plot(D['Time_s'], D['Channel_2_V'], label='Output')
pl.xlabel('Time (s)')
pl.ylabel('Voltage (V)')
pl.legend()
pl.title('Voltage Divider 3dB')

E.dtype
E['Time_s'], E['Channel_1_V'], E['Channel_2_V']
pl.figure()
pl.plot(E['Time_s'], E['Channel_1_V'], label='Input')
pl.plot(E['Time_s'], E['Channel_2_V'], label='Output')
pl.xlabel('Time (s)')
pl.ylabel('Voltage (V)')
pl.legend()
pl.title('Closed Loop with a Follower Wire')

F.dtype
F['Frequency_Hz'], F['Channel_1_Magnitude_dB'], F['Channel_2_Magnitude_dB'], 
pl.figure()
pl.plot(F['Frequency_Hz'], F['Channel_1_Magnitude_dB'], label='Input')
pl.plot(F['Frequency_Hz'], F['Channel_2_Magnitude_dB'], label='Output')
pl.xlabel('Frequency (Hz)')
pl.ylabel('Magnitude (dB)')
pl.legend()
pl.title('Op-Amp with 1k ohm Resistor Before Feedback Wire')

G.dtype
G['Frequency_Hz'], G['Channel_1_Magnitude_dB'], G['Channel_2_Magnitude_dB'],
pl.figure()
pl.plot(G['Frequency_Hz'], G['Channel_1_Magnitude_dB'], label='Input')
pl.plot(G['Frequency_Hz'], G['Channel_2_Magnitude_dB'], label='Output')
pl.xlabel('Frequency (Hz)')
pl.ylabel('Magnitude (dB)')
pl.legend()
pl.title('Op-Amp with 1k ohm Resistor After Feedback Wire')

H.dtype
H['Time_s'], H['Channel_1_V'], H['Channel_2_V']
pl.figure()
pl.plot(H['Time_s'], H['Channel_1_V'], label='Input')
pl.plot(H['Time_s'], H['Channel_2_V'], label='Output')
pl.xlabel('Time (s)')
pl.ylabel('Voltage (V)')
pl.legend()
pl.title('Non-Inverting Amplifier')

I.dtype
I['Frequency_Hz'], I['Channel_1_Magnitude_dB'], I['Channel_2_Magnitude_dB'], 
pl.figure()
pl.plot(I['Frequency_Hz'], I['Channel_1_Magnitude_dB'], label='Input')
pl.plot(I['Frequency_Hz'], I['Channel_2_Magnitude_dB'], label='Output')
pl.xlabel('Frequency (Hz)')
pl.ylabel('Magnitude (dB)')
pl.legend()
pl.title('Non-Inverting Amplifier Network Analysis')

J.dtype
J['Time_s'], J['Channel_1_V'], J['Channel_2_V']
pl.figure()
pl.plot(J['Time_s'], J['Channel_1_V'], label='Input')
pl.plot(J['Time_s'], J['Channel_2_V'], label='Output')
pl.xlabel('Time (s)')
pl.ylabel('Voltage (V)')
pl.legend()
pl.title('Inverting Amplifier')

K.dtype
G['Frequency_Hz'], G['Channel_1_Magnitude_dB'], G['Channel_2_Magnitude_dB'], 
pl.figure()
pl.plot(K['Frequency_Hz'], K['Channel_1_Magnitude_dB'], label='Input')
pl.plot(K['Frequency_Hz'], K['Channel_2_Magnitude_dB'], label='Output')
pl.xlabel('Frequency (Hz)')
pl.ylabel('Magnitude (dB)')
pl.legend()
pl.title('Inverting Amplifier Network Analysis')

```
