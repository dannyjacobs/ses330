```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
D = np.genfromtxt('Waveform_test.csv', comments='#', skip_header=19, names=True, delimiter=',')
```

```python
D.dtype
D['Time_s'], D['Channel_1_V']
pl.figure()
pl.plot(D['Time_s'], D['Channel_1_V'], label='Signal')
pl.plot(D['Time_s'], D['Channel_2_V'], label='Noise')
pl.xlabel('Time (s)')
pl.ylabel('Voltage (V)')
pl.legend()
```
