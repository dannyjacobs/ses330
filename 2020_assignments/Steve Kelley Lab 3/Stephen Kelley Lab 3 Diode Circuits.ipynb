```python
import numpy as np
import matplotlib.pyplot as pl
A = np.genfromtxt('3l_1_network_analysis.csv', comments='#', skip_header=20, names=True, delimiter=',')
B = np.genfromtxt('3l_1_resonance_freq_reversed.csv', comments='#', skip_header=20, names=True, delimiter=',')
C = np.genfromtxt('3l_1_resonance_sweep.csv', comments='#', skip_header=20, names=True, delimiter=',')
D = np.genfromtxt('lab3l_1_Skelleyresonance_freq_FFT.csv', comments='#', skip_header=20, names=True, delimiter=',')
E=np.genfromtxt('3l_2_half_wave.csv', comments='#', skip_header=28, names=True, delimiter=',')
F=np.genfromtxt('lab3l_2_Skelleyfull_bridge_rectifier.csv', comments='#', skip_header=28, names=True, delimiter=',')
G=np.genfromtxt('3l_3_full_wave_recitfier_ripple_22microf.csv', comments='#', skip_header=28, names=True, delimiter=',')
```

imports programming, and assigns variables to the data exported from the lab. 

```python
A.dtype
A['Frequency_Hz'], A['Channel_1_Magnitude_dB'], A['Channel_2_Magnitude_dB'], A['Channel_2_Phase_deg']
pl.figure()
pl.plot(A['Frequency_Hz'], A['Channel_1_Magnitude_dB'], label='Input')
pl.plot(A['Frequency_Hz'], A['Channel_2_Magnitude_dB'], label='Output')
pl.xlabel('Frequency (Hz)')
pl.ylabel('Attenuation (dB)')
pl.legend()
pl.title('Network Analysis of an LC Crcuit')

B.dtype
B['Frequency_Hz'], B['Channel_1_Magnitude_dB'], B['Channel_2_Magnitude_dB'], B['Channel_2_Phase_deg']
pl.figure()
pl.plot(B['Frequency_Hz'], B['Channel_1_Magnitude_dB'], label='Input')
pl.plot(B['Frequency_Hz'], B['Channel_2_Magnitude_dB'], label='Output')
pl.xlabel('Frequency (Hz)')
pl.ylabel('Attenuation (dB)')
pl.legend()
pl.title('Network Analysis of a Reverse LC Circuit')

C.dtype
C['Time_s'], C['Channel_1_V'], C['Channel_2_V']
pl.figure()
pl.plot(C['Time_s'], C['Channel_1_V'], label='Input')
pl.plot(C['Time_s'], C['Channel_2_V'], label='Output')
pl.xlabel('Time (s)')
pl.ylabel('Voltage (V)')
pl.legend()
pl.title('Resonance Over Sweeping Input Frequencies')

D.dtype
D['Time_s'], D['Channel_1_V'], D['Channel_2_V']
pl.figure()
pl.plot(D['Time_s'], D['Channel_1_V'], label='Input')
pl.plot(D['Time_s'], D['Channel_2_V'], label='Output')
pl.xlabel('Time (s)')
pl.ylabel('Voltage (V)')
pl.legend()
pl.title('Fourier Analysis of a Square Wave')

E.dtype
E['Time_s'], E['Channel_1_V'], E['Channel_2_V']
pl.figure()
pl.plot(E['Time_s'], E['Channel_1_V'], label='Input')
pl.plot(E['Time_s'], E['Channel_2_V'], label='Output')
pl.xlabel('Time (s)')
pl.ylabel('Voltage (V)')
pl.legend()
pl.title('Half Wave Rectifier Input vs. Output')

F.dtype
F['Time_s'], F['Channel_1_V'], F['Channel_2_V']
pl.figure()
pl.plot(F['Time_s'], F['Channel_1_V'], label='Input')
pl.plot(F['Time_s'], F['Channel_2_V'], label='Output')
pl.xlabel('Time (s)')
pl.ylabel('Voltage (V)')
pl.legend()
pl.title('Full Bridge Rectifier input vs. Output')

G.dtype
G['Time_s'], G['Channel_1_V'], G['Channel_2_V']
pl.figure()
pl.plot(G['Time_s'], G['Channel_1_V'], label='Input')
pl.plot(G['Time_s'], G['Channel_2_V'], label='Output')
pl.xlabel('Time (s)')
pl.ylabel('Voltage (V)')
pl.legend()
pl.title('Full Bridge Rectifier With Capacitor Ripple')
```
