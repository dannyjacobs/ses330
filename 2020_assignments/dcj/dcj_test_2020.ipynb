```python
import numpy as np
import matplotlib.pyplot as pl
```

Load data. Skip 19 lines of comments, use the actual column names from the file. 

```python
D = np.genfromtxt('dcj_test_save.csv', comments='#',skip_header=19, names=True, delimiter=',')
```

```python
D.dtype
```

```python
D['Time_s'], D['Channel_1_V']
```

```python
pl.figure()
pl.plot(D['Time_s'], D['Channel_1_V'], label='Input')
pl.plot(D['Time_s'], D['Channel_2_V'], label='Output')
pl.xlabel('time (s)')
pl.ylabel('voltage (v)')
pl.legend()
```
